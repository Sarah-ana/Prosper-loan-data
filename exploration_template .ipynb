# Univariate Analysis

In this analysis,We'll be answering the following question

1. What the Borrower’s interest rate for prosper loan
2. The Borrower’s APR for prosper loan. Every borrower is concerned about the APR, he/she will be charged for the loan. 
3. And lastly,does the borrower owns house at the time of listing (True & False)

Now let's our analysis answer the question.

```python
#creating a new dataframe
df = loan.copy()
```

```python
# borrower rate
def lables_name(x , y = None):
    plt.xlabel(x)
    plt.ylabel(y)
    
    
plt.figure(figsize = [10,15])
sns.distplot(df['BorrowerRate']);
lables_name('BorrowerRate')
plt.ylabel('Frequency') 
```

```python
df['BorrowerRate'].describe()
```

From the above analysis,we can detect that the distribution isn't normal. But, what we can bring out from this is that our mean is 0.19 and the maximum borrowed rate is 0.49 and the minimum is 0.0. And the standard deviation for this analysis is 0.07.


Now, let's look for the analysis for the burrower APR.

```python
plt.figure(figsize = [10,15])
sns.distplot(df['BorrowerAPR']);
lables_name('BorrowerAPR')
plt.ylabel('frequeny') 
```

```python
df['BorrowerAPR'].describe()
```

From the above analysis,every borrower is concerned about the APR, he/she will be charged for the loan, so now from the analysis  we can see that the average a person would be chargered for a loan is 0.2, the maximum would be 0.51 and the minimum would be 0.00. So, the mean of this analysis is 0.21 and the standard deviation of this analysis is 0.08.



Now, a question came to mind,and the question is, how many of this borrower are home owners?
Let's use analysis to answer that.

```python
base_color = sns.color_palette()[0]
sns.countplot(data = df , x='IsBorrowerHomeowner' , color =base_color)
plt.ylabel('counts of home owner') 
lables_name('IsBorrowerHomeowner')
```

```python
df['IsBorrowerHomeowner'].describe()
```

So,now has this analysis answwered your question?
From this analysis we can see that it's balanced, which implies that borrowers that have their own home are half and burrowers that does't have their own home are half.

# Bivariate Exploration
In this analysis, we would be finding the relationship between two entities,but,what question would be considered.
Let's consider the following question 
1. What the current type of employment of prosper rating
2. How worthy are the burrowers
3. Within the last 6 months how many borrowers make more frequent inquiries?
4. What is the APR for risky borrowers
5. How can we be sure that high risk borrowers must have higher BankCardUtilizaiton percentage.

Now let's consider those questions 


```python
#Relationships between EmploymentStatust vs ProsperRating (Alpha)
plt.figure(figsize = [20,15])
sns.countplot(x= 'EmploymentStatus', hue = 'ProsperRating (Alpha)',data = df)
plt.show;
```

```python
df['EmploymentStatus'].describe()
```

So, i wanted to know what was the relationship between employment status and prosper rating because the rating assigned by Prosper helps to decide the APR the borrower should be charged.So, from the analysis we can see those that are employed would be charged higher than self-employed,full-time,part-time,retired and others. So, under the employed AA would be the least charged while the highest charged would be C. 

```python
#Relationships between CreditGrade vs BorrowerRate
plt.figure(figsize = [15,10])
sns.boxplot(y = 'CreditGrade', x= 'BorrowerAPR',data = df)

plt.show;
```

I wanted to analyze credit grade,because  is important to judge the credit worthiness of a borrower, and from the above  analysis we can judge that all the borrower are worthy to borrow. 

```python
#Relationships between InquiriesLast6Months vs BorrowerRate
plt.figure(figsize = [15,10])
sns.scatterplot(y = 'InquiriesLast6Months', x= 'BorrowerAPR',data = df)
plt.ylabel('InquiriesLast6Months')
plt.show
```

```python
df['InquiriesLast6Months'].describe()
```

From the above scatter plot analysis we can see that within the last 6 month, we noticed that the borrowers who make more frequent inquiries would we charged higher APR, but we really don't have much difference.

```python
#Relationships between ProsperRating (Alpha) vs BorrowerRate
plt.figure(figsize = [15,10])
sns.violinplot(y = 'BorrowerRate', x= 'ProsperRating (Alpha)',data = df)
plt.ylabel('ProsperRating (Alpha) ')
plt.show
```

From this analysis,we can say that the APR for risky borrowers is significantly higher as compared to low risk borrowers

```python
#Relationships between ProsperRating (Alpha) vs BankcardUtilization
plt.figure(figsize = [15,10])
sns.boxplot(y = 'BankcardUtilization', x= 'ProsperRating (Alpha)',data = df)
plt.ylabel('BankcardUtilization')
plt.show
```

Since ProsperRating(Alpha) is a categorical variable, boxplot is a suitable choice,from the above analysis,we can say the distribution of BankcardUtilization is left skewed, because risky borrowers might have utilized their cards to the limit.So, we can’t be sure that HighRisk borrowers must have higher BankcardUtilizaiton percentage.

# Multivariate Exploration

In this analysis, we are going to be considering the following question 

1.  Is it possible for a  borrowers to overspend and hence would need additional funds?
2. Would the size of loan amount between terms be bigger.

Let consider this 

```python
#LoanOriginalAmount,IncomeRange and CreditGrade relationship
def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn """
    default_color = sns.color_palette()[0]
    sns.boxplot(x, y, color = default_color)
g=sns.FacetGrid(data= df, aspect=1.2, height=5, col='IncomeRange', col_wrap=4)
g.map(sns.boxplot, 'CreditGrade', 'LoanOriginalAmount');
g.add_legend()
```

From the above boxplot, all I can say is that the income range of the average loan amount requested is lower for risky borrowers. It is expected for risky borrowers to get lower amount approved because of their low credit worthiness.It's risky because borrowers do have outliers across every Income range. One possible explanation is that these borrowers tend to overspend and hence need additional funds.

```python
g=sns.FacetGrid(data= df, aspect=1.2, height=5, col='Term', col_wrap=4)
g.map(sns.pointplot, 'ProsperRating (Alpha)', 'LoanOriginalAmount');
g.add_legend()
```

From the above analysis, I noticed that the loan amount of all three terms increases, the increase size of loan amount between terms also becomes larger.
